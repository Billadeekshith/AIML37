```python
people = ['luffy', 'zoro', 'sanji', 'nami', 'robin', 'brook', 'usoff', 'franky','jembaie','garp', 'sabo', 'ace']

age = [20,24,26,22,27,35,19,20,26,49,20,21]

weight = [49,40,46,48,50,62,35,69,18,85,48,48]

height = [160,135,170,165,173,168,175,159,105,171,155,158]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter (weight, height)
plt.title("relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")

plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.title("stock shares in daily life", fontsize=16, fontstyle='italic')

# Main plot function 'bar'

plt.bar(x=people, height=weight, width=0.6, color='blue', edgecolor='k', alpha=0.6)

plt.xlabel("People", fontsize=20)

plt.xticks(fontsize=14, rotation=25)

plt.yticks(fontsize=17)

plt.ylabel("stock (in prices)",fontsize=18)

plt.show()
```

```python
import numpy as np

plt.figure(figsize=(7,5))

# Main plot function 'hist'

plt.hist(weight,color='red', edgecolor='k', alpha=0.75,bins=5)

plt.title("my life time income", fontsize=15)

plt.xlabel("increase of income",fontsize=15)

plt.xticks(fontsize=15)

plt.yticks (fontsize=15)

plt.show()

```
